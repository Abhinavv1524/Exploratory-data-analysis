```python
df['Bed Count'] = df['Number of bed'].str.split(' ').str[0]
```

```python
df['Bed Type'] = df['Number of bed'].str.split(' ').str[1].replace({'beds':'Normal'})
```

```python
df.drop('Number of bed', axis = 1, inplace=True)
```

```python
df.head(2)
```

```python
df['Header'] = df['Title'].str.split(',').str[0]
df['State'] = df['Title'].str.split(',').str[1]
df['Country'] = df['Title'].str.split(',').str[2]
```

```python
df.drop('Title', axis=1, inplace=True)
```

```python
df.head()
```

```python
df['Check in Month'] = df['Date'].str.split(' ').str[0]
df['Check in Date'] = df['Date'].str.split(' ').str[1]
df['Check out Month'] = df['Date'].str.split(' ').str[0]
df['Check out Date'] = df['Date'].str.split(' ').str[3]

```

```python
df.drop('Date', axis=1, inplace=True)
```

```python
df.head()
```

```python
df['House Type'] = df['Header'].str.split(' in').str[0]
df['Location'] = df['Header'].str.split(' in').str[1]
```

```python
df.drop('Header', axis=1, inplace=True)
```

```python
df.head()
```

```python
df['Rating'] = df['Review and rating'].str.split('(').str[0]
df['Review'] = df['Review and rating'].str.split('(').str[1].str.split(')').str[0]
```

```python
df.drop('Review and rating', inplace=True, axis=1)
```
